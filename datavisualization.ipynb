```python
import cv2
import dlib
import os

# Initialize dlib's face detector
detector = dlib.get_frontal_face_detector()

def crop_face_from_video(input_video_path, output_video_path):
    # Open the input video
    cap = cv2.VideoCapture(input_video_path)
    if not cap.isOpened():
        print(f"Error: Unable to open video {input_video_path}")
        return

    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Detect faces in the frame
        faces = detector(frame)

        if len(faces) > 0:
            # Assume the first detected face is the main character
            face = faces[0]

            # Get the coordinates of the face rectangle
            x, y, w, h = face.left(), face.top(), face.width(), face.height()

            # Ensure the face coordinates are within frame bounds
            x = max(0, x)
            y = max(0, y)
            w = min(w, frame_width - x)
            h = min(h, frame_height - y)

            if w > 0 and h > 0:
                # Crop the face from the frame
                cropped_face = frame[y:y+h, x:x+w]

                # Resize the cropped face to maintain the output video size
                cropped_face_resized = cv2.resize(cropped_face, (frame_width, frame_height))

                # Write the cropped face to the output video
                out.write(cropped_face_resized)

        frame_count += 1
        print(f"Processing frame {frame_count}/{total_frames}")

    # Release the video capture and writer objects
    cap.release()
    out.release()
    print(f"Finished processing {input_video_path}, saved to {output_video_path}")

# Directory containing videos
input_directory = '/content/drive/MyDrive/Colab Notebooks/Celeb-DF/9_deleted_videos'
output_directory = '/content/drive/MyDrive/Colab Notebooks/Celeb-DF/9_deleted_videos_cropped'
os.makedirs(output_directory, exist_ok=True)

# Process each video in the directory
for video_filename in os.listdir(input_directory):
    if video_filename.endswith(('.mp4', '.avi', '.mov')):
        input_video_path = os.path.join(input_directory, video_filename)
        output_video_path = os.path.join(output_directory, f"{video_filename}")
        crop_face_from_video(input_video_path, output_video_path)
```

```python
import os

def remove_prefix_from_filenames(directory, prefix):
    # List all files in the specified directory
    for filename in os.listdir(directory):
        # Check if the filename starts with the specified prefix
        if filename.startswith(prefix):
            # Define the old and new file paths
            old_filepath = os.path.join(directory, filename)
            new_filename = filename[len(prefix):]  # Remove the prefix
            new_filepath = os.path.join(directory, new_filename)

            # Rename the file
            os.rename(old_filepath, new_filepath)
            print(f"Renamed: {old_filepath} -> {new_filepath}")

# Directory containing the output videos
output_directory = '/content/drive3/MyDrive/Colab Notebooks (1)/Celeb-DF/All_videos_croped3'

# Prefix to be removed
prefix_to_remove = 'cropped_'

# Call the function to remove the prefix from filenames
remove_prefix_from_filenames(output_directory, prefix_to_remove)
```

```python
import os
import shutil

def move_videos(src_folder, fake_folder, real_folder):
    # Create destination folders if they don't exist
    os.makedirs(fake_folder, exist_ok=True)
    os.makedirs(real_folder, exist_ok=True)

    # Loop through all files in the source folder
    for filename in os.listdir(src_folder):
        if filename.lower().endswith(('.mp4', '.avi', '.mov', '.mkv')):  # Add other video extensions as needed
            file_path = os.path.join(src_folder, filename)

            # Check if "id" appears twice in the filename
            if filename.count("id") >= 2:
                shutil.move(file_path, os.path.join(fake_folder, filename))
                print(f'Moved "{filename}" to "{fake_folder}"')
            else:
                shutil.move(file_path, os.path.join(real_folder, filename))
                print(f'Moved "{filename}" to "{real_folder}"')

# Define source and destination folders
src_folder = '/content/drive3/MyDrive/Colab Notebooks/Celeb-DF/All_videos_croped'
fake_folder = '/content/drive3/MyDrive/Colab Notebooks/Celeb-DF/fake_videos_croped'
real_folder = '/content/drive3/MyDrive/Colab Notebooks/Celeb-DF/real_videos_croped'

# Call the function
move_videos(src_folder, fake_folder, real_folder)
```

```python
def get_video_length(video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error opening video file: {video_path}")
        return None

    # Get the total number of frames
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Get the frame rate (frames per second)
    frame_rate = int(cap.get(cv2.CAP_PROP_FPS))

    # Calculate video length in seconds
    video_length = total_frames / frame_rate

    # Release the video capture object
    cap.release()

    return video_length

def get_frame_size(frame_path):
    # Read a sample frame
    frame = cv2.imread(frame_path)
    if frame is None:
        print(f"Error reading frame: {frame_path}")
        return None

    # Get frame dimensions (height, width)
    height, width, _ = frame.shape

    return height, width
```

```python
def calculate_average_fps(folder_path):
    total_fps = 0
    total_videos = 0

    for filename in os.listdir(folder_path):
        if filename.lower().endswith((".mp4", ".avi", ".mkv")):
            video_path = os.path.join(folder_path, filename)
            cap = cv2.VideoCapture(video_path)
            fps = cap.get(cv2.CAP_PROP_FPS)
            total_fps += fps
            total_videos += 1
            cap.release()

    if total_videos > 0:
        average_fps = total_fps / total_videos
        return average_fps
    else:
        return None


# Calculate average FPS for real and fake videos
average_real_fps = calculate_average_fps(realdata)
average_fake_fps = calculate_average_fps(fakedata)

if average_real_fps:
    print(f"Average FPS for real videos: {average_real_fps:.2f}")
else:
    print("No real videos found.")

if average_fake_fps:
    print(f"Average FPS for fake videos: {average_fake_fps:.2f}")
else:
    print("No fake videos found.")
```

```python
frames_list = []
def frames_extraction(video_path):
    '''
    This function will extract the required frames from a video after resizing and normalizing them.
    Args:
        video_path: The path of the video in the disk, whose frames are to be extracted.
    Returns:
        frames_list: A list containing the resized and normalized frames of the video.
    '''

    # Declare a list to store video frames.

    SEQUENCE_LENGTH = 20

    # Read the Video File using the VideoCapture object.
    video_reader = cv2.VideoCapture(video_path)

    # Get the total number of frames in the video.
    video_frames_count = int(video_reader.get(cv2.CAP_PROP_FRAME_COUNT))

    # Calculate the the interval after which frames will be added to the list.
    skip_frames_window = max(int(video_frames_count/SEQUENCE_LENGTH), 1)

    # Iterate through the Video Frames.
    for frame_counter in range(SEQUENCE_LENGTH):

        # Set the current frame position of the video.
        video_reader.set(cv2.CAP_PROP_POS_FRAMES, frame_counter * skip_frames_window)

        # Reading the frame from the video.
        success, frame = video_reader.read()

        # Check if Video frame is not successfully read then break the loop
        if not success:
            break

        # Resize the Frame to fixed height and width.
        resized_frame = cv2.resize(frame, (224, 224))

        # Normalize the resized frame by dividing it with 255 so that each pixel value then lies between 0 and 1
        normalized_frame = resized_frame / 255

        # Append the normalized frame into the frames list
        frames_list.append(normalized_frame)

    # Release the VideoCapture object.
    video_reader.release()

    # Return the frames list.
    return frames_list

folder_path = "/content/drive/MyDrive/Colab Notebooks (1)/Celeb-DF/fake_data"
for file_name in os.listdir(folder_path):
    video_path = os.path.join(folder_path, file_name)
    frames_list = frames_extraction(video_path)
    print(frames_list)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def highlight_face_and_blur_background(frame, face_cascade):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    mask = np.zeros_like(frame)
    for (x, y, w, h) in faces:
        mask = cv2.rectangle(mask, (x, y), (x+w, y+h), (255, 255, 255), -1)

    blurred_frame = cv2.GaussianBlur(frame, (10001, 10001), 0)
    highlighted_frame = np.where(mask == np.array([255, 255, 255]), frame, blurred_frame)

    return highlighted_frame

# Read the image
image_path = '/content/drive/MyDrive/Colab Notebooks (1)/02_07__meeting_serious__1JCLEEBQ.mp4_984.jpg'
frame = cv2.imread(image_path)

# Check if the image is loaded properly
if frame is None:
    raise FileNotFoundError(f"Image file not found at {image_path}")

# Process the image
result = highlight_face_and_blur_background(frame, face_cascade)

# Display the result
cv2_imshow(result)


```
