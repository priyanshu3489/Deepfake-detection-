```python
import cv2
import os

def get_video_length(video_path):
    """Returns the length of the video in seconds."""
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Cannot open video file {video_path}")
        return None
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    cap.release()
    if fps > 0:
        return frame_count / fps
    else:
        return None

def get_videos_length(folder_path):
    """Prints the length of all videos in the folder and calculates the average length.
       Also prints the video file names and IDs if the video length is greater than 20 seconds."""
    video_extensions = ('.mp4', '.avi', '.mov', '.mkv', '.flv', '.wmv')  # Add more extensions if needed
    lengths = []
    video_ids = []
    for idx, file_name in enumerate(os.listdir(folder_path)):
        if file_name.lower().endswith(video_extensions):
            video_path = os.path.join(folder_path, file_name)
            length = get_video_length(video_path)
            if length is not None:
                lengths.append(length)
                print(f"Video: {file_name}, Length: {length:.2f} seconds")
                if length > 15:
                    video_ids.append((idx, file_name))

    if lengths:
        avg_length = sum(lengths) / len(lengths)
        print(f"Average Length: {avg_length:.2f} seconds")
    else:
        print("No valid video files found.")

    if video_ids:
        print("\nVideos longer than 15 seconds:")
        for video_id, video_name in video_ids:
            print(f"ID: {video_id}, Video: {video_name}")


if __name__ == "__main__":
    folder_path = "/content/drive/MyDrive/Colab Notebooks/Celeb-DF/fake_data"  # Change this to your folder path
    get_videos_length(folder_path)
```
